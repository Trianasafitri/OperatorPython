# Operator Aritmatika

```python
apel = 12
teman = 4
hasil_bagi = apel / teman 
print("Setiap teman dapat:", hasil_bagi)

apel+= 8
print("Total apel sekarang:", apel)
```

# Operator Perbandingan

```python
siti = 160 
andi = 165

if siti > andi:
    print("Siti lebih tinggi")
elif siti < andi:
    print("Andi lebih tinggi")
else:
    print("Tinggi mereka sama")
```

# Operator Logika

```python
cuaca_cerah = True
sudah_selesai_pr = True 

if cuaca_cerah and sudah_selesai_pr:
    print("Budi bisa bermain")
else:
    print("Budi tidak bisa bermain")
```

# Operator Bitwise

```python
a = 6   #0110
b = 3   #0011

print("AND:", a & b)
print("OR:", a | b)
print("XOR:", a ^ b)
```

# Operator Penugasan

```python
saldo = 50000
saldo += 20000   # isi ulang
saldo -= 30000   # beli paket

print("Sisa saldo:", saldo)
```

# Operator Keanggotaan

```python
peserta = ["Andi", "Budi", "Citra", "Dewi"]

print("Eka" in peserta)

kalimat = "Saya suka belajar Python"
print("Python" in kalimat)
```

# Operator Identitas

```python
x = 10
y = 10

print(x is y)

list1 = [1,2,3]
list2 = [1,2,3]

print(list1 is list2)
```

# Operator Ternary

```python
angka = 120
print("Lebih besar dari 100" if angka > 100 else "Tidak")

nilai = 75
print("Lulus" if nilai > 70 else "Tidak Lulus")
```
